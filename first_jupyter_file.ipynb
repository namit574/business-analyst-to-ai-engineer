IMPORTANT CONCEPT — LOCAL COMMITS
Right now your commit exists ONLY on your computer.
You have NOT uploaded to GitHub yet.

create git ignore so that no noise will be uploaded to github
touch .gitignore

WHY .gitignore MATTERS A LOT
Without .gitignore:
Git may upload:
virtual environments
cache
secrets
huge unnecessary files
Bad practice.

You only setup Git globally once on your machine.
For every new project, you mainly need to:

Create project folder
Create new virtual environment (venv)
Initialize git in that folder
Connect that folder to a new GitHub repo

Typical flow:

# Create project folder
mkdir my_project
cd my_project

# Create virtual environment
python -m venv venv

# Activate venv
# Windows
venv\Scripts\activate

# Mac/Linux
source venv/bin/activate

Now setup Git for THIS project:

git init

Create .gitignore file:

venv/
__pycache__/
.ipynb_checkpoints/
.env

Then:

git add .
git commit -m "Initial commit"
Create GitHub repo

Go to GitHub

Create a new repository like:

my_project

Then connect local project to GitHub:

git remote add origin https://github.com/yourusername/my_project.git
git branch -M main
git push -u origin main

After this, normal workflow is:

git add .
git commit -m "message"
git push
Important Understanding
Git vs GitHub
Git → version control on your laptop
GitHub → online storage for git repositories

You install/configure Git once globally:

git config --global user.name "Your Name"
git config --global user.email "you@example.com"

After that, every project only needs:

git init
Recommended Structure
Projects/
│
├── project1/
│   ├── venv/
│   ├── src/
│   └── .git/
│
├── project2/
│   ├── venv/
│   ├── app/
│   └── .git/

Each project:

has its own venv
has its own git repo
can connect to separate GitHub repo
Super Common Beginner Mistake

Do NOT do:

One venv for all projects
One git repo for all projects

Keep everything isolated project-wise.
