2) row based --->>

df.loc[0]           # first row (label-based)

Two Ways Pandas Accesses Data
1. 🔢 iloc → Index-based (position-based)
df.iloc[0]

👉 Means:
“Give me the row at position 0 (first row), no matter what the label is.”

Works like Python lists
Always uses integer positions: 0, 1, 2, 3...
2. 🏷️ loc → Label-based
df.loc[0]

👉 Means:
“Give me the row whose index label is 0”

Uses index names/labels, not position
Labels can be anything: numbers, strings, dates

### Group By 
In pandas, groupby is most commonly used with aggregation, but it’s not limited to it.

Typical use:

df.groupby("dept")["salary"].mean()

This collapses rows → one result per group.

🔹 2. Can we do row-level (window-like) operations?

Yes—you can, but not with .agg().

For row-level results (like SQL window functions), you use:

✅ transform() (closest to window functions)
df["dept_avg_salary"] = df.groupby("dept")["salary"].transform("mean")
in sql :-  AVG(salary) OVER (PARTITION BY dept)

👉 This keeps the same number of rows as the original DataFrame
👉 Each row gets its group’s aggregate value

This is very similar to:

AVG(salary) OVER (PARTITION BY dept)




3. ROW_NUMBER() OVER (PARTITION BY dept ORDER BY salary)
SQL
ROW_NUMBER() OVER (PARTITION BY dept ORDER BY salary)

Pandas
df["row_number"] = df.sort_values(["dept", "salary"]) \
                     .groupby("dept") \
                     .cumcount() + 1

```python
import pandas as pd

df = pd.DataFrame({
    "dept": ["A", "A", "A", "B", "B"],
    "emp":  ["e1","e2","e3","e4","e5"],
    "salary": [10, 20, 30, 40, 50]
})
```

## Handling Missing Data

In real datasets, you’ll constantly see missing values like:

NaN
None

```python
df.isna().sum() #this is perfect to see how many null values per column we have this is perfect to check in large dataset
```

Merging & Joining DataFrames

👉 Used when you have multiple tables (datasets) and want to combine them.

Think like Excel VLOOKUP or SQL JOIN.

Joining datasets using pd.merge(on = " " , how = " " )

for joininin we have

1) Merge - Merge two datasets in inner join , left join , right join , outer join like in SQL 

2) Contact - stacking two datasets obove one another 

```python
df1 = pd.DataFrame({
    "CustomerID": [1, 2, 3],
    "Name": ["Amit", "Riya", "Raj"]
})



df2 = pd.DataFrame({
    "CustomerID": [1, 2, 4],
    "OrderAmount": [500, 700, 300]
})
```

```python
#we will first use Merge if we want to merge another dataset in main then we use df1.merge(df2...) 
# we can also use pd.merge(df1,df2)

#lets check both 
```

```python
df1.merge(df2,on = 'CustomerID') # merged like inner join 
```

```python
#lets try second syntax

pd.merge(df1,df2) #automatically chose column column and did inner join
```

```python
pd.merge(df1,df2, how = 'left') #left join mean all the columns from df1 and only matching from df2
```

```python
pd.merge(df1,df2,how = 'right') #right join mean all the columns from df1 and only matching from df2
```

```python
pd.merge(df1,df2 , how = 'outer') #outer join keep all the rows non matching values will NaN
```

Merge on Different Column Names

when column names are different either we can rename the columns and join 
or we can do pd.merge(df1,df1,left_on = ' ' , right_on = ' ' )

```python
df3 = pd.DataFrame({
    "CustomerID": [1, 2, 3],
    "Name": ["Amit", "Riya", "Raj"]
})



df4 = pd.DataFrame({
    "CustID": [1, 2, 4],
    "OrderAmount": [500, 700, 300]
})
```

```python
pd.merge(df3,df4,left_on = 'CustomerID',right_on = 'CustID') #inner join based on different column names we can use how = '' for left right
```

```python
pd.merge(df3,df4,left_on = 'CustomerID',right_on = 'CustID',how = 'left') #inner join based on different column names, using how 
```

```python
# 2 ) concat

# 1. The "List" Rule
# 2 The Axis Rule (Direction) axis = 0 , which is defailt and 


pd.concat([df1,df2]) #we need to pass dataframes into lists
```

```python
pd.concat([df3,df4]) #column names are different  hence side by side 
```
