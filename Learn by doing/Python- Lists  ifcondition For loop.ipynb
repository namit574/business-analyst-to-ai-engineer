FOR LOOP AHEAD
for loop is used when you want to iterate through something like list or string 

```python
expense = [1000,1200,1500,1800,1200]
```

```python
for i in expense: #i will go to every elemnet insise expense list and then there is a print statement which prints i
    print(i)
```

```python
total_expenses = 0
for i in expense:
    total_expenses = total_expenses+i #everytime i will go to each element and instead of print we add it to total_expenses
print(total_expenses) 
```

```python
for i in range(len(expense)): #this is how we print month which is basically index of elements in list and then list element
    expenses = expense[i]
    print(f"Month:{i+1} expense:{expenses}")
```

 BREAK STATEMENT IN FOR LOOP = WHEN WE USE BREAK IT WILL CLOSE THE LOOP AS IS AND COME OUT OF THE LOOP

```python
sales = [34,33,28,43,56,34,39,29]
```

```python
thresold = 33
```

```python
for sale in sales:
    if sale<thresold:
        print(f"sale {sale} is less then thresold") # if you see loop has gone all the way to the end of the list
```

```python
#we use break statement when first sales below thresold we find 
```

```python
for sale in sales:
    if sale<thresold:
        print(f"sale {sale} is less then thresold")
        break
    
```
