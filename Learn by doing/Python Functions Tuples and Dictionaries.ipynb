## Pass by value and position 
### we usually say call by value function or call by reference
#this is nothing but used when we interchange the position of parameters 

```python
def sum_of_two(a,b):
    total = a+b
    return total #here we have used return statement instead of print which mean when we call the fucntion whatever is in
#return will be returned or printed
```

```python
sum_of_two(3,4) # here 3 is a and 3 is b by default - which is call by reference
```

```python
sum_of_two(b=4,a=7) #here we have interchanged by we have given them values - which is call by value
```

### we have different types of arguments


    1) Positional or Required Arguments --- explained above

    2) Keyword Arguments --- like name age variables etc

    3) Default Arguments ---- we fix the default value to a 

    4) Positional-only Arguments ---- you cant interchange the arguments by defining argument variables - def(a,b) , def(b=3,a=4) this you cant do , positional only defined using \ ex:- def(a,b \):

    5) Keyword-only arguments -- keyword-only arguments are function parameters that must be supplied by name (e.g., key=value) during a function call, rather than by their position. They are defined by placing a bare asterisk (*) in the parameter list—everything after the * must be named. def make_box(width, height, *, color="red", filled=False):
    print(f"{width}x{height} {color} box. Filled: {filled}")

Correct usage
make_box(10, 20, color="blue", filled=True)

Incorrect usage (will raise TypeError)
make_box(10, 20, "blue", True)

       6) Arbitrary or Variable-length Arguments def name (*args): a function that accepts variable number of arguments which becomes tuples and you can either itetrate over them or do tuple operations
       
       def name (*kargs) stands for keyword arguments - difference between *args bs *krwgs is *args convert arguments in tuple but *kwrgs converts into key value pairs
    


```python
def evenodd():
    num = input("Please enter a number to check even or odd")
    num = int(num)
    if num%2 ==0:
        print("Number is even")
    else:
        print("Number is odd")
```

```python
evenodd()
```

```python
def max_of_num(num1,num2):
    if num1 > num2:
        return num1
    else:
        return num2
```

```python
max_of_num(5,4)
```
