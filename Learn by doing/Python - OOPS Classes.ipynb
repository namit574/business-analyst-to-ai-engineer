# Derived class or inheritance 
when another class inherits all the properties of parent class
syntax

class BaseClass:
    # Body of the base class (attributes and methods)
    pass

class DerivedClass(BaseClass):
    # Body of the derived class (additional attributes and methods,
    # or overrides of base class methods)
    pass
    
    
    
1️⃣ What happens when you inherit

Suppose you have a parent class LibraryItem:

class LibraryItem:
    def __init__(self, title, price):
        self.title = title
        self.price = price
        self.is_borrowed = False

Every object of LibraryItem has title, price, and is_borrowed.

Now you make a child class:

class Book(LibraryItem):
    def __init__(self, author):
        self.author = author
2️⃣ Will it inherit attributes automatically?

NO — here’s why:

When you define __init__ in the child class, Python does NOT automatically call the parent’s __init__.

That means self.title, self.price, and self.is_borrowed will NOT exist unless you explicitly call the parent constructor.

book = Book("John Doe")
print(book.title)   # ❌ AttributeError

You’ll get an error because title was never initialized.

3️⃣ Correct way: use super()
class Book(LibraryItem):
    def __init__(self, title, price, author):
        super().__init__(title, price)  # call parent constructor
        self.author = author

super().__init__(title, price) calls the parent’s constructor, so self.title, self.price, and self.is_borrowed are properly set.

Then you can add child-specific attributes like self.author.

Usage:

book1 = Book("Kenny the Book", 500, "John Doe")
print(book1.title)       # Kenny the Book
print(book1.price)       # 500
print(book1.author)      # John Doe
print(book1.is_borrowed) # False

✅ Works perfectly.

4️⃣ Summary
Scenario	What happens
Child has no __init__	Inherits everything from parent, including parent’s __init__
Child defines its own __init__ but does not call super()	Parent attributes not initialized, parent constructor is skipped
Child defines its own __init__ and calls super().__init__()	Parent attributes initialized, child attributes added → best practice

So the key rule:

If your child class defines __init__, you must call super().__init__() if you want parent attributes to exist.

->> BASICALLY OF YOU WANT TO HAVE ALL THE ATTRIBUTES FROM PARENT CLASS AS WELL AS YOU WANT TO ADD SOME MORE ATTRIBUTES IN CHILD CLASS USE SUPER().__INIT__()
