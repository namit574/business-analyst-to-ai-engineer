# Exception Handling 

Inside a block of code when some error occurs below code does not run. but we want to run below code hence we use exception handling.

Syntax:--
try:
    some code where error can occur
except:
    if above code has some error this part of code will run 

```python
#example when a number is divided by 0 we get error div error

num1 = int(input("enter first number"))
num2 = int(input("enter second number"))
d = 0 
try:
    d = num1/num2
    print(d)
except:
    print (f"exception raised and handeled d = {d}")
```

```python
balance = 0 

def deposit(amount):
    global balance
    balance +=amount

deposit(50)
print(f"Now Balance is {balance}")
```

```python
def withdrawal(amount):
    global balance
    if amount>balance:
        raise Exception ("withdrawal amount is more than balance") #exeption using raise 
    else:
        balance -=amount
        print(f"remaining amount is {balance}")
        print("hello customer what are you doing man")
        
        
print(balance)     
withdrawal(20)
```

Code written immediately after a
raise statement within the same block will not execute because raise stops the current execution and transfers control to an exception handler.
