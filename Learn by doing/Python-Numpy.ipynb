# why numpy is fast and how memory allocation is done ?

Numpy is fast due to :
1) continuous memory allocation because it stores same datatypes 
but in list we can put many data types and it doesnot store in continous memory locations but in random 

2) while searching in np array whole array is scanned at once however in list a loop works


🎯 Why NumPy is Fast

Because it uses:

C language implementation

Vectorized operations ---- Instead of loops, NumPy processes the entire array at once.

Continuous memory layout

No Python loops --

Broadcasting - NumPy can operate on arrays of different sizes.

The numpy.arange() function (commonly referred to as np.arange()) is a powerful tool in Python's NumPy library used to create an array with evenly spaced values within a specified interval. It functions similarly to Python's built-in range() function but returns a NumPy ndarray

```python
np.arange(4,8)
```

```python
#Que What is the shape of an array created using np.zeros((3, 4))?

x = np.zeros((3,4))
print(x.shape)
```
